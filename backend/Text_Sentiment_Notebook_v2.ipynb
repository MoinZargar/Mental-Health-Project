```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

stopwords_set = set(stopwords.words('english'))
emoji_pattern = re.compile('(?::|;|=)(?:-)?(?:\)|\(|D|P)')

def preprocessing(text):


    # Remove HTML tags
    text = re.sub('<[^>]*>', '', text)

    # Remove URLs
    text = re.sub(r'http\S+', '', text)

    # Extract emojis
    emojis = emoji_pattern.findall(text)

    # Remove non-alphanumeric characters, convert to lowercase, and join emojis
    text = re.sub('[\W+]', ' ', text.lower()) + ' '.join(emojis).replace('-', '')



   # Apply stemming and remove stopwords
    prter = PorterStemmer()
    text = [prter.stem(word) for word in text.split() if word not in stopwords_set]

    processed_text= " ".join(text)
    return processed_text


```

```python
# Define a mapping dictionary
label_mapping = {'sadness':0, 'joy':1, 'love':2, 'anger':3, 'fear':4,'surprise':5}

# Rename the 'Label' column using the mapping dictionary
data['label'] = data['label'].map(label_mapping)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
# TF-IDF Vectorization
tfidf=TfidfVectorizer(strip_accents=None,lowercase=False,preprocessor=None,use_idf=True,norm='l2',smooth_idf=True,max_features=1000)
x = tfidf.fit_transform(data.text).toarray()
y = data.label.values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
# Reshape the input data for LSTM
# LSTM expects input in the shape of (samples, time steps, features)
# In this case, samples would be the number of documents, time steps would be 1 (since each document is treated independently), and features would be the TF-IDF values
x_train_lstm = np.reshape(x_train, (x_train.shape[0], 1, x_train.shape[1]))
x_test_lstm = np.reshape(x_test, (x_test.shape[0], 1, x_test.shape[1]))
```

```python
x_train_lstm.shape
```

```python
from keras.layers import Input

# Define input layer explicitly
input_layer = Input(shape=(1, x_train.shape[1]))

# Define LSTM model
model = Sequential()
model.add(input_layer)
model.add(LSTM(units=128))
model.add(Dense(units=6, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

print(model.summary())
```

```python
### Finally Training
# Train the model
model.fit(x_train_lstm,y_train,validation_data=(x_test_lstm,y_test),batch_size=64,epochs=10)
```

```python
# Evaluate the model
score = model.evaluate(x_test_lstm, y_test)
print('Test Accuracy:', score[1])
```

```python
def preprocessing2(text):


    # Remove HTML tags
    text = re.sub('<[^>]*>', '', text)

    # Remove URLs
    text = re.sub(r'http\S+', '', text)

    # Extract emojis
    emojis = emoji_pattern.findall(text)

    # Remove non-alphanumeric characters, convert to lowercase, and join emojis
    text = re.sub('[\W+]', ' ', text.lower()) + ' '.join(emojis).replace('-', '')
    # lower case conversion
    text=text.lower()

    # Tokenization, remove stopwords and lemmatization


    lemmatizer=WordNetLemmatizer()
    words=word_tokenize(text)
    pos_tags = nltk.pos_tag(words)
    lemmatized_words = [lemmatizer.lemmatize(word, pos=getPOS(pos)) for word, pos in pos_tags if word.lower() not in set(stopwords.words('english'))]

    processed_text=' '.join(lemmatized_words)
    print("Processed Text:",processed_text)
    oneHotEncodedText=oneHotEncoding(processed_text)
    return oneHotEncodedText
```

```python
import numpy as np

def analyze_sentiment(comment):
    print("Comment:", comment)
    sentiments = []

    # Preprocess the comment and convert it to one-hot encoded text
    oneHotEncodedText = preprocessing2(comment)

    # Get the decision scores from the model
    decision_scores = model.predict(oneHotEncodedText)
    print("Raw Decision Scores: [Negative, Positive, Neutral]", decision_scores)

    # Get the indices of top 3 predicted classes
    top_3_indices = np.argsort(-decision_scores[0])[:3]  # Select the top 3 indices
    print("Top 3 predicted class indices:", top_3_indices)

    # Sum of probabilities of the top 3 sentiments
    sum_probabilities = np.sum(decision_scores)

    # Map the predicted class indices to sentiments
    label_mapping = {0: 'sadness', 1: 'joy', 2: 'love', 3: 'anger', 4: 'fear', 5: 'surprise'}
    for index in top_3_indices:
        sentiment = label_mapping[int(index)]
        probability = decision_scores[0][int(index)]  # Access the probability corresponding to the index
        percentage = (probability / sum_probabilities) * 100
        sentiments.append((sentiment, probability, percentage))

    return sentiments




```
